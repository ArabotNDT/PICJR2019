```python
import rospy
from std_msgs.msg import String
```

```python
def enviar():
     pub = rospy.Publisher('correio', String, queue_size=10)
     rospy.init_node('remetente', anonymous=True)
     rate = rospy.Rate(10) # 10hz
     while not rospy.is_shutdown():
         hello_str = "Nova mensagem %s" % rospy.get_time()
         rospy.loginfo(hello_str)
         pub.publish(hello_str)
         rate.sleep()
```

```python
enviar()
```
